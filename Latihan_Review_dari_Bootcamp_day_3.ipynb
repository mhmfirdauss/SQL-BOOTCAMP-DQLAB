```python
# Import library yang dibutuhkan
from google.colab import auth, data_table
from google.cloud import bigquery
from pandas_gbq import to_gbq

# Proses autentikasi akun
auth.authenticate_user()
print('Authenticated')
```

```python
# Buat BigQuery client
project_id = 'cobabelajar-123'
client = bigquery.Client(project = project_id)
```

```python
%%bigquery --project {project_id} --verbose

Management ingin mengetahui seluruh data transaksi di toko, sehingga tampilkan semua data pada tabel orders!
-- Tampilkan semua data pada tabel orders

select *
FROM cobabelajar-123.perusahaan_db.Penjualan
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan semua data no_urut as nomor, cash_type, money, coffee_name, Date, time dan urutkan transaksinya

SELECT
  no_urut as nomor,
  cash_type,
  money,
  coffee_name,
  Date,
  time
FROM cobabelajar-123.perusahaan_db.Penjualan
ORDER BY nomor ASC

```

```python
%%bigquery --project {project_id} --verbose


-- Tampilkan kolom no_urut as nomor, cash_type, money, coffee_name, Date, time pada tabel penjualan
-- namun ubah penamaan coffee_name pada output menjadi tipe_kopi

SELECT
  No_urut as nomor,
  cash_type,
  money,
  coffee_name as tipe_kopi,
  Date,
  time
FROM cobabelajar-123.perusahaan_db.Penjualan
ORDER BY nomor ASC
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan dari ada apa aja tipe kopi yang di beli pada tabel Pembelian

SELECT DISTINCT
  coffee_name as Tipe_Kopi
FROM cobabelajar-123.perusahaan_db.Penjualan
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan apa saja tipe pembayaran kopi yang sering dipakai dan
-- Jenis kopi apa saja yang di pesan

SELECT DISTINCT
  cash_type as Tipe_Pembayaran,
  coffee_name as Tipe_Kopi
FROM cobabelajar-123.perusahaan_db.Penjualan
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan semua 5 data pada tabel pembelian

SELECT *
FROM cobabelajar-123.perusahaan_db.Penjualan
LIMIT 5
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan 10 data teratas saja untuk kolom cash_type, coffee_name , Date pada tabel penjualan

SELECT
  cash_type ,
  coffee_name ,
  Date
FROM cobabelajar-123.perusahaan_db.Penjualan
LIMIT 10
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan 5 data teratas saja untuk kolom cash_type, coffee_name , Date pada tabel penjualan
-- namun ubah penamaan cash_type, coffee_name pada output menjadi tipe_pembayaran , Tipe_kopi

SELECT
  Cash_type as tipe_pembayaran,
  coffee_name as Tipe_kopi,
  Date
FROM cobabelajar-123.perusahaan_db.Penjualan
LIMIT 5
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan semua data pada tabel penjualan terurut berdasarkan kolom no_urut secara ascending

SELECT *
  FROM cobabelajar-123.perusahaan_db.Penjualan
  ORDER BY no_urut ASC
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan semua data pada tabel penjualan terurut berdasarkan kolom date secara descending

SELECT *
FROM cobabelajar-123.perusahaan_db.Penjualan
ORDER BY
  Date DESC
```

```python
%%bigquery --project {project_id} --verbose

-- Tampilkan nama kopi yang dipesanan dan waktu pagi / siang / malam yang pernah terjadi,
-- Urutkan berdasarkan nama kopi secara descending dan waktu pagi siang dan malam secara ascending!

SELECT
  coffee_name as Nama_Kopi,
  Time_of_day as Waktu,
  DATE as Tanggal
  FROM cobabelajar-123.perusahaan_db.Penjualan
  ORDER BY
    Nama_Kopi DESC,
    Waktu ASC
```
